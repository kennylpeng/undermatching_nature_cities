```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
from create_program_df_export import create_program_df, add_offer_rates
from create_counterfactuals_export import create_counterfactual_df
from create_student_df_export import create_student_df
from create_extended_student_df_export import create_extended_student_df
from create_filtered_program_df_export import create_programs_with_apps_df
```

```python
df_students = create_student_df(2022, include_ms_tests=False, include_enrollment=False)
df_programs = create_program_df(2022, include_ms_tests=False)
df_programs = add_offer_rates(df_programs, df_students)
df_counterfactuals, programs_counterfactual = create_counterfactual_df(df_students, df_programs)
programs_counterfactual = df_counterfactuals.columns.to_list()
df_programs['counterfactual'] = df_programs['programcode'].apply(lambda x: x in programs_counterfactual)
df_students_extended = create_extended_student_df(df_students, df_programs, df_counterfactuals, programs_counterfactual, drop_helpers=False, impute_clusters=True)
df_programs_with_apps = create_programs_with_apps_df(df_programs, df_students_extended)
```

```python
df_programs.to_csv('../clean_dfs/programs_export.csv')
df_students_extended.to_csv('../clean_dfs/students_extended_export.csv')
df_programs_with_apps.to_csv('../clean_dfs/programs_with_apps_export.csv')
```
